```python
import pandas.io.sql as sqlio
import psycopg2 as ps
import os
from dotenv import load_dotenv
import pandas as pd
import numpy as np
```

```python
import pandas as pd
import os
import numpy as np
import datetime as dt
from datetime import date, time, datetime
import math
```

```python
load_dotenv()
db = os.getenv('POSTGRES_DB')
user = os.getenv('POSTGRES_USER')
password = os.getenv('POSTGRES_PASSWORD')
host = os.getenv('POSTGRES_SERVER')
port = os.getenv('POSTGRES_PORT')
```

```python
conn = ps.connect(dbname=db, 
                  user=user,
                  password=password, 
                  host=host,
                  port=port)
```

```python
# Accommodate query
sql = """SELECT id_port, code, port_name
FROM stg_pelni.master_port
where  flag_ppss like 'Y'"""
```

```python
# Read query
df = sqlio.read_sql_query(sql, conn)
df
```

```python
# Check columns name
df.columns.values
```

```python
# Loc columns
df = df.loc[:, ['id_port', 'code','port_name']]
df
```

```python
# add columns and dat in new columns
df['nautical'] = '99999'
df['commercial'] = '99999'
```

```python
# Add new columns, but data from column 'id_port'
df['port_origin'] = df['id_port']
df['port_destination'] = df['id_port']

df['code_origin'] = df['code']
df['code_destination'] = df['code']
```

```python
# Loc columns
df = df.loc[:, ['code_origin', 'code_destination', 'port_origin', 'port_destination', 'nautical', 'commercial']]
```
